# For Loop practice

```python
#1.Print a string using for loop

x=['apple', 'orange', 'banana', 'mango']

for i in range(len(x)):
    print(x[i],i)

#If I want to access specific index what should I do?

#1.Print first 20 integer

for x in range(1, 20, 1):
    print(x+1)
    
#2.Sum of first 100 positive integer

total = 0

for i in range(1, 101):
      total += i

print(total)
```

# While loop

```python
i=1
sum=0
while i < 101:
    
    sum=sum+i   
    i = i+1
    
    
print(sum)   
```
